```python
class Demo:
    def __init__(self, param1):
        self.param1=param1
    def __str__(self,):
        return "We are understanding str function"
    
    def __repr__(self):
        return f"Demo(param1='{self.param1}')"
```

```python
d = Demo("value1")
```

```python
d
```

```python
Demo(param1='value1')
```
